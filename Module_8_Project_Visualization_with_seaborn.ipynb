```python
#import necessary libraries

import matplotlib
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as npy
import seaborn as sns 
```

```python
#Loading the datset into the dataframe
df = pd.read_csv('WorldCupMatches.csv ')
print(df.head())
```

```python
#Visualizing the total number of goals scored in each match
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df.head())
```

```python
#Total number of goals scored in each Match
plt.figure(figsize=(12,8))
sns.displot(df['Total Goals'] , bins = 20, kde=False)
plt.title('Total Number of Goals in each Match')
plt.savefig('Total Number of Goals in each Match.png')
plt.show()
```

```python
#Creating a figure with size of the figure to be 12 inches wide and 7 inches tall.
sns.set_style('whitegrid')
sns.set_context(rc={'font.size': 10, 'xtick.labelsize':8})
f, ax = plt.subplots(figsize=(12, 7))
ax = sns.barplot(data=df, x='Year', y='Total Goals', ci=None, estimator=sum)
plt.xticks(rotation='vertical')
ax.set_title('Goals Scored Each Year Between 1930-2014')
plt.savefig('Goals Scored Each Year Between 1930-2014.png')
plt.show()
```

```python
#creating a box plot to visualize the distribution of the average goals for each year with a bar chart.
sns.set_context('notebook', font_scale=1.25)
f, ax2 = plt.subplots(figsize=(12, 7))
sns.color_palette('Spectral')
ax2 = sns.boxplot(data=df, x= 'Year', y='Total Goals')
plt.xticks(rotation='vertical')
ax2.set_title('Goals per year')
plt.savefig('Goals per Year.png')
plt.show()
```
